```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Unemployment in India.csv")
```

Data Preprocessing

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={' Date': 'Date'}, inplace=True)
df.rename(columns={' Frequency': 'Frequency'}, inplace=True)
df.rename(columns={' Estimated Unemployment Rate (%)': 'Estimated Unemployment Rate (%)'}, inplace=True)
df.rename(columns={' Estimated Employed': 'Estimated Employed'}, inplace=True)
df.rename(columns={' Estimated Labour Participation Rate (%)': 'Estimated Labour Participation Rate (%)'}, inplace=True)
```

```python
df['Frequency'] = df['Frequency'].str.strip()
df['Frequency'].unique()
```

```python
sns.boxplot(df['Estimated Unemployment Rate (%)'])
```

Analyse statistique

```python
df['Estimated Unemployment Rate (%)'].describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Estimated Unemployment Rate (%)'], bins=30, kde=True)
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Number of observation')
plt.title('Distribution of unemployment rate')
```

l'histo montre une asymetrie positive  (Moyenne > Mediane)

```python
df['Estimated Unemployment Rate (%)'].mean()
```

```python
df['Estimated Unemployment Rate (%)'].median()
```

**Estimated Unemployment Rate (%) by Date**

```python
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values("Date", inplace=True)

plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Estimated Unemployment Rate (%)'], marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Evolution of the unemployment rate over time')
plt.xticks(rotation=45)
plt.grid(True)
```

Une augmentation du taux de chomage pendant COVID-19

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Region", y="Estimated Unemployment Rate (%)", data=df)
plt.xticks(rotation=90)
plt.title("Comparaison du taux de chômage par région")
plt.show()
```

**Estimated Unemployment Rate (%) by Region**

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=df, hue='Region', palette='viridis')

plt.title('Unemployment Rate by Region', fontsize=16)
plt.xlabel('Region', fontsize=14)
plt.ylabel('Unemployment Rate (%)', fontsize=14)

plt.xticks(rotation=90)
plt.tight_layout()
```

**Estimated Unemployment Rate (%) by Area**

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Area', y='Estimated Unemployment Rate (%)', data=df, hue='Area', palette='viridis')

plt.title('Unemployment Rate by Area', fontsize=16)
plt.xlabel('Area', fontsize=14)
plt.ylabel('Unemployment Rate (%)', fontsize=14)

plt.xticks(rotation=90)
plt.tight_layout()
```
